```python
import pandas as pd
```

```python
df=pd.read_csv('play.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
outlook=LabelEncoder()
```

```python
df['outlook']=outlook.fit_transform(df['outlook'])
df['humidity']=outlook.fit_transform(df['humidity'])
df['windy']=outlook.fit_transform(df['windy'])
df['play']=outlook.fit_transform(df['play'])
```

```python
df
```

```python
x=df[['outlook','humidity','windy']]
```

```python
x
```

```python
y=df[['play']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
reg=DecisionTreeClassifier(criterion='gini')
```

```python
reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```
